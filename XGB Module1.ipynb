```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
import warnings
warnings.filterwarnings('ignore')
```

```python
data=pd.read_csv(r'cleaned_data.csv')
data
```

```python
x=data.drop(['y'],axis=1)
x
```

```python
y=data[['y']]
y
```

```python
data.drop('euribor3m',axis=1,inplace=True)
```

```python
data.drop('duration',axis=1,inplace=True)
```

```python
data
```

```python
from sklearn.preprocessing import MinMaxScaler
sc=MinMaxScaler()
```

```python
x_scaled=sc.fit_transform(x)
x_scaled
```

```python
x=pd.DataFrame(x_scaled,columns=x.columns)
x
```

```python
# pip install --upgrade scikit-learn imbalanced-learn
```

```python
from imblearn.over_sampling import SMOTE
from collections import Counter
sm=SMOTE()
```

```python
print(Counter(y))
x_sm,y_sm=sm.fit_resample(x,y)
print(Counter(x_sm))
print(Counter(y_sm))

```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_sm,y_sm,test_size=0.25,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from xgboost import XGBClassifier
```

```python
xgb=XGBClassifier()
```

```python
xgb.fit(x_train,y_train)
```

```python
xgb_predict=xgb.predict(x_test)
xgb_predict
```

```python
from sklearn.metrics import classification_report
# Assuming y_test and xgb_predict are defined

# Generate classification report
report = classification_report(y_test, xgb_predict)

# Print the classification report
print(report)

```

```python
import pandas as pd
import numpy as np

# Flatten xgb_predict if it's a 2D array
xgb_predict_flat = xgb_predict.flatten() if len(xgb_predict.shape) > 1 else xgb_predict

# Convert y_test to a numpy array and then flatten it
y_test_flat = y_test.to_numpy().flatten() if len(y_test.shape) > 1 else y_test.to_numpy()

# Create crosstab
crosstab_result = pd.crosstab(y_test_flat, xgb_predict_flat)

# Print the result
print(crosstab_result)

```

```python
from sklearn.model_selection import RandomizedSearchCV
```

```python
n_estimators=[int(x) for x in np.linspace(start=200,stop=2000,num=10)]
max_depth=[2,3,4,5]
learning_rate=[0.1,0.2,0.3]
gamma=[0,1,2,3,4]
subsample=[0.5,0.7,1]
colsample_bytree=[0.5,0.7,1]
```

```python
params={'max_depth':max_depth,'learning_rate':learning_rate,'n_estimators':n_estimators,
       'gamma':gamma,'subsample':subsample,'colsample_bytree':colsample_bytree}
```

```python
XGB=XGBClassifier(random_state=42)
```

```python
rcv=RandomizedSearchCV(XGB,scoring='accuracy',param_distributions=params,
                      n_iter=100,cv=3,random_state=42,n_jobs=-1)
```

```python
rcv.fit(x_train,y_train)
```

```python
cv_best_params=rcv.best_params_
```

```python
print(f"best parameters:{cv_best_params}")
```

```python
cv_best_params
```

```python
XGB2=XGBClassifier(subsample=0.7,
 n_estimators=400,
 max_depth=5,
 learning_rate=0.1,
 gamma=0,
 colsample_bytree=0.7)
```

```python
XGB2.fit(x_train,y_train)
```

```python
y_predict_xgb=XGB2.predict(x_test)
```

```python
y_predict_xgb
```
